```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print(" Minimum recharge amount is ₹50")
        return False
    if amount not in valid_plans:
        print(" Invalid plan. Choose from:", valid_plans)
        return False
     return True
# Retry loop
while True:
    try:
        amt = int(input("Enter recharge amount: "))
         if validate_recharge(amt):
            print("Recharge successful!")
            break
        else:
            print("Please try again.\n")
     except ValueError:
        print(" Please enter a valid number.\n")
```

```python
def check_reorder(products):
    for item, qty in products.items():
        if qty < 15:
            print(f"{item}: Reorder Alert (Stock = {qty})")
        else:
            print(f"{item}: Stock OK (Stock = {qty})")

# Example dictionary
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 14,
    "Salt": 25
}

check_reorder(inventory)
```

```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m
    average = total / len(marks)
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    return average, result

student_marks = [60, 45, 70, 55]
avg, res = process_result(student_marks)
print("Average Marks:", avg)
print("Result:", res)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)
    return fare
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        total_fare = calculate_fare(km, peak_hour)
        print(f" Total Fare: ₹{total_fare:.2f}")
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using the fare estimator ")
            break
    except ValueError:
        print(" Please enter valid distance.\n")
```

Problem 5:

```python
def check_eligibility(attendance):
    present_days = 0
    for day in attendance:
        if day.upper() == "P":
            present_days += 1
    percentage = (present_days / len(attendance)) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
emp_attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
perc, result = check_eligibility(emp_attendance)
print(f"Attendance Percentage: {perc:.2f}%")
print("Status:", result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd = input("Enter password: ")
result = check_password_strength(pwd)
print("Result:", result)
```
